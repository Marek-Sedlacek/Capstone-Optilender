### Calculate Month of Loss
And create loss at 12M Flag

```python
def loan_payment(loan, ir, n):
    payment = loan * ((ir * ((1+ir) ** n)) / ((1+ir) ** n - 1))
    return payment
```

```python
loan_payment(18000,0.1373/12,60)
```

```python
def loss_month_approx(loan, ir, n, act_principal_paid,mob):
    try:
        pmt = loan_payment(loan,ir,n)
        current_bal = loan
        principal_paid = 0
    
        for i in range(1,n+1):
            current_bal = current_bal*(1+ir)-pmt
            principal_paid = loan - current_bal
    #         print principal_paid, act_principal_paid
            if principal_paid > act_principal_paid: break
            elif abs(principal_paid/act_principal_paid-1) <0.05: break

        if i>mob: return -1
        else: return i#, principal_paid/act_principal_paid-1
    except:
        return -1
```

```python
Loss["stopped_payment"] = Loss.apply(lambda x: loss_month_approx(x['amount_borrowed'], 
                                                          x['borrower_rate']/12,
                                                         x['term'],
                                                         x['principal_paid'],
                                                         x['age_in_months']), axis=1)

Loss["loss_month"] = Loss["stopped_payment"]+3
Loss["loss_month"][Loss["stopped_payment"]==-1] =-1
```

## Expected Return at MOB 12

```python
def interest_approx(loan, ir, n, act_principal_paid, act_interest_paid):
    
    prepayment_rate = 0.1
    
    try:
        pmt = loan_payment(loan,ir,n)
        current_bal = loan
        exp_principal_paid = 0
        exp_interest_paid = 0       
    
        for i in range(1,13):
            current_bal = current_bal*(1+ir)-pmt
            
            exp_principal_paid = loan - current_bal
            exp_interest_paid = pmt*12 - exp_principal_paid
#             print exp_interest_paid
        
        return exp_interest_paid*(1-prepayment_rate)
    
    except:
        return -1
```
